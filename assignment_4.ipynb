# Program to scrape most viewed videos on youtube


```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import time
```

```python
url = 'http://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos'
driver = webdriver.Chrome()
driver.get(url)
```

```python
table = driver.find_element(By.XPATH,'//*[@id="mw-content-text"]/div[1]/table[2]/tbody')
Rank = []
Name = []
Artist = []
Upload_Date = []
Views = []
```

```python
rows = table.find_elements(By.TAG_NAME,'tr')
```

```python
for row in rows:
    cols = row.find_elements(By.TAG_NAME,'td')
    if len(cols)>0:
        Rank.append(cols[0].text)
        Name.append(cols[1].text)
        Artist.append(cols[2].text)
        Upload_Date.append(cols[3].text)
        Views.append(cols[4].text)
```

```python
df = pd.DataFrame({'Rank' : Rank,'Name' : Name,'Artist' : Artist,'Upload_Date' : Upload_Date,'Views' : Views})
df
```

# Program To Scrape The Details of Highest Selling Novals

```python
url = 'https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare'
driver = webdriver.Chrome()
driver.get(url)

```

```python
Book_Name = []
Author_Name = []
Volumes_Sold= []
Publisher = []
Genre = []
```

```python
table = driver.find_element(By.XPATH,'//*[@id="article-body-blocks"]/div/table/tbody') 
rows = table.find_elements(By.TAG_NAME,'tr')
```

```python
for row in rows:
    cols = row.find_elements(By.TAG_NAME,'td')
    Book_Name.append(cols[1].text)
    Author_Name.append(cols[2].text)
    Volumes_Sold.append(cols[3].text)
    Publisher.append(cols[4].text)
    Genre.append(cols[5].text)
```

```python
df = pd.DataFrame({'Book Name':Book_Name,'Author Name':Author_Name,'Volumes Sold':Volumes_Sold,'Publisher':Publisher,'Genre':Genre})
df
```

# Program to scrape deatails of datasets from uci


```python
url = 'https://archive.ics.uci.edu/'
driver = webdriver.Chrome()
driver.get(url)
```

```python
all_data_sets = driver.find_element(By.XPATH,'/html/body/table[1]/tbody/tr/td[2]/span[2]/a')
all_data_sets.click()
```

```python
table = driver.find_element(By.XPATH,'/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody') 
rows = table.find_elements(By.TAG_NAME,'tr')
```

```python
dataset_name = []
data_type = []
task = []
attribute_type = []
no_of_instances = []
no_of_attributes = []
year = []
```

```python
for row in rows[1:]:
    cols = row.find_elements(By.TAG_NAME,'td')
    dataset_name.append(cols[1].text)
    #data_type.append(cols[2].text)
```

```python
Type = driver.find_elements(By.XPATH,'//table[@border="1"]/tbody/tr/td[2]/p')
for i in Type[1:]:
    if i.text is None:
        data_type.append('--')
    else :
        data_type.append(i.text)
```

```python
tas = driver.find_elements(By.XPATH,'//table[@border="1"]/tbody/tr/td[3]/p')
for i in tas[1:]:
    if i.text is None:
        task.append('--')
    else:
        task.append(i.text)
```

```python
attribute = driver.find_elements(By.XPATH,'//table[@border="1"]/tbody/tr/td[4]/p')
for i in attribute[1:]:
    if i.text is None :
        attribute.append('--')
    else:
        attribute_type.append(i.text)
```

```python
instances = driver.find_elements(By.XPATH,'//table[@border="1"]/tbody/tr/td[5]/p')
for i in instances[1:]:
    if i.text is None:
        no_of_instances.append('--')
    else:
        no_of_instances.append(i.text)
```

```python
attri = driver.find_elements(By.XPATH,'//table[@border="1"]/tbody/tr/td[6]/p')
for i in attri[1:]:
    if i.text is None :
        no_of_attributes.append('--')
    else:
        no_of_attributes.append(i.text)
    
```

```python
y = driver.find_elements(By.XPATH,'//table[@border="1"]/tbody/tr/td[7]/p')
for i in y[1:]:
    if i.text is None:
        year.append('--')
    else:
        year.append(i.text)
```

```python
df = pd.DataFrame({'Dataset Name':dataset_name,'Data Type':data_type,'Task':task,'Attribute Type':attribute_type,'No Of Instances':no_of_instances,'No Of Attributes':no_of_attributes,'Year':year})
df
```
