```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
url = 'https://en.wikipedia.org/wiki/Main_Page'

response = requests.get(url)

soup = bs(response.text, 'html.parser')

header_tags = []
header_tags_text = []

for header in soup.find_all(['h1', 'h2', 'h3', 'h4', 'h5', 'h6']):
    header_tags.append(header.name)
    header_tags_text.append(header.get_text())

header_dict = {'Header': header_tags, 'Text': header_tags_text}

df = pd.DataFrame(header_dict)

print(df)
```

```python
from bs4 import BeautifulSoup 
import requests
import pandas
url = 'https://www.imdb.com/chart/top/'

response = requests.get(url)

soup = BeautifulSoup(response.text, 'html.parser')

movie_titles = []
movie_ratings = []
movie_years = []

movie_data = soup.select('td.titleColumn')
movie_ratings_data = soup.select('td.posterColumn span[name="ir"]')

for i in range(len(movie_data)):
    title = movie_data[i].find('a').get_text()
    year = movie_data[i].find('span', class_='secondaryInfo').get_text()
    rating = movie_ratings_data[i]['data-value']
    
    movie_titles.append(title)
    movie_years.append(year)
    movie_ratings.append(rating)

movie_dict = {'Title': movie_titles, 'Rating': movie_ratings, 'Year': movie_years}

df = pd.DataFrame(movie_dict)

print(df.head(50))
```
